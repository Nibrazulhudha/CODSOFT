Separating X and Y

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

**Model Selection**

Train_test_split:Evaluate the performance of a model on unseen data. It involves dividing dataset into training set and testing set.

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
#Training data=70% & testing data=30%
```

Random state-setting random state=42 ensures reproducability and consistent results in ML algorithms that involve randomness.

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
#Used to transform the variables of a dataset to a common range.
#Range varies between 0 & 1.
```

```python
x_train
```

**Algorithms**

```python
from sklearn.tree import DecisionTreeClassifier
#Creating object
dec=DecisionTreeClassifier()
dec.fit(x_train,y_train)
#Predicting using the test data
y_pred=dec.predict(x_test)
#Calculating accuracy
from sklearn.metrics import accuracy_score,classification_report
print("Decision Tree Accuracy:",accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

**Sample Prediction**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report
rf=RandomForestClassifier(n_estimators=100,random_state=42)
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
#Calculating Accuracy
print("Random Forest Accuracy:",accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

**Confusion Matrix**

A confusion matrix is an N x N matrix that is used to evaluate the performance of a classification model, where N is the number of target classes. It compares the actual target values against the ones predicted by ML model

```python
from sklearn.metrics import confusion_matrix
matrix=confusion_matrix(y_test,y_pred)
print(matrix)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
label=['Iris-versicolor','Iris-virginica','Iris-setosa']
cmd=ConfusionMatrixDisplay(matrix,display_labels=label)
cmd.plot()
```
